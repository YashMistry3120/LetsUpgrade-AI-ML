```python
import pandas as pd 
```

```python
dataset = pd.read_csv('general_data_copy.csv')
```

```python
dataset.info()
```

```python
dataset
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.drop_duplicates(inplace = True)
```

```python
dataset
```

```python
dataset.Attrition = dataset.Attrition.map({'Yes' : 1, 'No' : 0})
```

```python
dataset
```

```python
from sklearn import preprocessing
```

```python
Label_Encoder = preprocessing.LabelBinarizer()
```

```python
dataset["Gender"] = Label_Encoder.fit_transform(dataset["Gender"])
```

```python
dataset
```

```python
from scipy.stats import chi2_contingency

'H0 - There is no dependency between Attrition and Age (REJECTED)'
'H1 - There is dependency between Attrition and Age'

chitable = pd.crosstab(dataset.Attrition, dataset.Age)
```

```python
chitable
```

```python
stats1, p1, dof1, expected1 = chi2_contingency(chitable)
print(stats1, p1)


if p1 < 0.05 :
    print("We reject the Ho")
else:
    print("We Accept the Ho")
# in the result below we can say that we reject the Ho due to p < 0.05
```

```python
'''Ho:There is no dependency between BusinessTravel and Attribution
H1:There is dependency between BusinessTravel and Attribution'''

chitable = pd.crosstab(dataset.Attrition, dataset.BusinessTravel)
chitable
```

```python
stats1, p1, dof1, expected1 = chi2_contingency(chitable)
print(stats1, p1)


if p1 < 0.05 :
    print("We reject the Ho")
else:
    print("We Accept the Ho")
# in the result below we can say that we reject the Ho due to p < 0.05
```
