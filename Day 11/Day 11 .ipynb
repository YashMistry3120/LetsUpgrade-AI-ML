```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df = pd.read_csv('S:\LetsUpgrade\Day 7\Assignment\general_data_copy.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Attrition'].replace(to_replace=("No","Yes"),value=(0,1),inplace=True)
```

```python
df = df.dropna()
df = df.drop_duplicates()
df
```

```python
df.corr()
```

```python
from scipy.stats import pearsonr
#p_value < 0.05 Reject H0
#p_value > 0.05 Do Not Reject H0 or Accepted
```

### Correlation Between Attrition and Age

```python
stats,p = pearsonr(df.Attrition,df.Age)
print("Correlation: ",stats,"\nP-Value: ",p)

if p<0.05:
    print("\n\nReject Null Hypothises")
else:
    print("\n\nAccept Null Hypothises")
```

```python
SampleData = df.sample(30)
plt.scatter(SampleData.Attrition,SampleData.Age)
```
